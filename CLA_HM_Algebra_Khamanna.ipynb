```python
import numpy as np
from matplotlib import pyplot as plt
from numpy.linalg import norm
import cv2
```

```python
img = cv2.imread('1632958775901.jpg')
print(img.shape)
```

```python
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img_transposed = img.T
print(img_transposed.shape)
plt.show()
```

```python
height = img[:,:,0].T
width = img[:,:,1].T
diag = img[:,:,2].T

img_t = np.dstack((height,width,diag))
plt.imshow(img_t)
plt.show
```

```python
# if it is diagonal or not.
 
N = np.array(height)
 
def isDiagonal(matrix) :
    for i in range(0, N):
        for j in range(0, N) :
             

            if ((i != j) and
             (matrix[i][j] != 0)) :
                return False
 
    return True

 
if (isDiagonal(img_t)) :
    print("Yes")
else :
    print("No")
```

```python
#If it's Scalar or not
M = width
def is_scalar(matrix) :
     
    for i in range(0,M) :
        for j in range(0,M) :
            if ((i != j)
               and (matrix[i][j] != 0)) :
                return False

    for i in range(0,M-1) :
        if (matrix[i][i] !=
           matrix[i + 1][i + 1]) :
            return False
 
    return True

if (is_scalar(img_t)):
    print("Yes")
else :
    print("No")
```
