```python
#Write a Python function that checks whether a passed string is palindrome or not.
def if_palindrome(my_str):
    rev_str = my_str[::-1]
    if my_str == rev_str:
        print("The string is palindrome.")
    else:
        print("The string is not palindrome.")
```

```python
#check
print(if_palindrome("anna"))
if_palindrome("annam")
```

```python
#Write a Python function that takes a number as a parameter and checks if the number is prime or not.
def is_prime(num):
    if num > 1:  
        for n in range(2,num):  
            if (num%n) == 0:  
                return False
        return True
    else:
        return False
```

```python
#check
print(is_prime(67))
print(is_prime(60))
```

```python
#Write a Python function to check whether a number is in a given range.
def in_range(n,x,m):
    if n in range(x,m):
        print( " %s is there"%str(n))
    else :
        print("The num is not there")
in_range(5,6,9)
```

```python
#Write a Python function to calculate the factorial of a number (a non-negative integer). 
#The function accepts the number as an argument.
def calculate_factorial(n):
    if n == 0:
        return 1
    else:
        return n * calculate_factorial(n-1)
```

```python
#check
calculate_factorial(7)
```

```python
#Write a Python program to reverse a string
def reverse_string(my_str):
    rev_str = my_str[::-1]
    return rev_str
```

```python
#check
reverse_string("nice weather")
```

```python
#Write a Python function to find the Max of three numbers.
def max_(a,b,c):
    max_nums = max(a,b,c)
    return max_nums
```
